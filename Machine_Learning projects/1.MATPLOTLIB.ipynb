```python
n = np.array([0,1,2,3,4,5])
fig,axes=plt.subplots(1,4,figsize=(16,5),dpi = 200)
axes[0].set_title("Scatter")
axes[0].scatter(x,x+0.25*np.random.randn(len(x)))
axes[1].set_title("step")
axes[1].step(n,n**2,lw=2)

axes[2].set_title("bar")
axes[2].bar(n,n**2,align="center",alpha=0.5)

axes[3].set_title("fill_between")
axes[3].fill_between(n,n**2,n**3,color="green",alpha=0.5)
```

```python
#Contour map
import matplotlib
import numpy as np
import matplotlib.cm as cm #cm for color map
import matplotlib.pyplot as plt


delta = 0.025
x=np.arange(-3.0,3.0,delta)
y = np.arange(-2.0,2.0,delta)
X,Y=np.meshgrid(x,y)
Z1 = np.exp(-X**2-Y**2)
Z2 = np.exp(-(X-1)**2-(Y-1)**2)
Z= (Z1-Z2)*2
```

```python
fig,ax=plt.subplots()
CS = ax.contour(X,Y,Z)
ax.clabel(CS,inline=1,fontsize=10)
ax.set_title("counter map")
```
