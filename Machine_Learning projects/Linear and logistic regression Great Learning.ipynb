```python
df = df.drop("make",axis=1)
df = df.drop("fuel-type",axis=1)
df = df.drop("engine-location",axis = 1)
df = df.drop("num-of-doors",axis=1)
df = df.drop("body-style",axis =1)
df = df.drop("drive-wheels",axis=1)
df = df.drop("engine-type",axis=1)
df = df.drop("fuel-system",axis =1)
df = df.drop("aspiration",axis =1)
df = df.drop("normalized-losses",axis=1)
```

```python
df.dtypes
```

```python
# one hot ecoding for gender
# lebel encodeer-orders the value not to be used with ordinal data
df["cylinder"]= df["num-of-cylinders"].replace({"one":1,"two":2,"three":3,"four":4,"five":5,"six":6,"seven":7})
df
```

```python
df['bore'] = pd.to_numeric(df['bore'],errors = 'coerce')
```

```python
# df['bore'] = pd.to_numeric(df['bore'],errors = 'coerce')
# df['stroke'] = pd.to_numeric(df['stroke'],errors = 'coerce')
# df['horsepower'] = pd.to_numeric(df['horsepower'],errors = 'coerce')
df['peak-rpm'] = pd.to_numeric(df['peak-rpm'],errors = 'coerce')
df['price'] = pd.to_numeric(df['price'],errors = 'coerce')
df['cylinder'] = pd.to_numeric(df['cylinder'],errors = 'coerce')
```

```python
df.info()
```

```python
#replacing missing values with median
df["price"] = df["price"].fillna(df["price"].median())
df["horsepower"] = df["horsepower"].fillna(df["horsepower"].median())
df["peak-rpm"] = df["peak-rpm"].fillna(df["peak-rpm"].median())
df["city-mpg"] = df["city-mpg"].fillna(df["city-mpg"].median())
df["highway-mpg"] = df["highway-mpg"].fillna(df["highway-mpg"].median())
df["price"] = df["price"].fillna(df["price"].median())
df["cylinder"] = df["cylinder"].fillna(df["cylinder"].median())
df["stroke"] = df["stroke"].fillna(df["stroke"].median())
df["bore"] = df["bore"].fillna(df["bore"].median())
```

```python
from sklearn.model_selection import train_test_split
array = pima_df.values
X=array[:,0:7]
y = array[:,8]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state = 7)
type(X_train)
```

```python
# fit the model on 30%
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
y_predict = model.predict(X_test)


model_score = model.score(X_test,y_test)
print(model_score)
print(metrics.confusion_matrix(y_test,y_predict))
```
