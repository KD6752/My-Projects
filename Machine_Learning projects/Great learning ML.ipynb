```python
conditions = [(df["MILES*"]<5),
              (df["MILES*"]>=5) & (df["MILES*"]<=10),
              (df["MILES*"]>10)]
values = ["Long Trip","Medium Trip","Short Trip"]

df["new_column"]= np.select(conditions,values)
df
```

```python
sf_cols = [col_name for col_name in housing.columns if "SF" in col_name]
sf_cols
```

```python
fig,axes = plt.subplots(nrows =3,ncols = 3,figsize = (10,10))
for i in range(0,len(sf_cols)):
    rows = i//3
    cols = i%3
ax= axes[rows,cols]
plot = sns.regplot(x=sf_cols[i],y = "SalePrice", data =housing,ax=ax)
```
